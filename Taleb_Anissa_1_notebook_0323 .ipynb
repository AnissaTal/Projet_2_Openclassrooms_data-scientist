 ###  <a name="C43">Analyse</a>


```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)#retire le calcul scientifique

listIndicatorChosen= ['Population growth (annual %)','Population of the official age for secondary education, both sexes (number)','Percentage of enrolment in post-secondary non-tertiary education in private institutions (%)', 'Government expenditure in educational institutions as % of GDP (%)','GDP per capita (current US$)','GDP at market prices (current US$)','Internet users (per 100 people)'] 

statsData=statsData.loc[statsData['Indicator Name'].isin(listIndicatorChosen) & statsData['Country Name']]
df=statsData.groupby(['Country Name','Indicator Name'])['value'].mean().to_frame().reset_index() # meme chose avec median et ecart type
#ecart type
dfSTD=statsData.groupby(['Country Name','Indicator Name'])['value'].std().to_frame().reset_index() # meme chose avec median et ecart type
dfSTD=dfSTD.set_index(['Country Name','Indicator Name']).unstack()['value'].reset_index()
#mediane
dfmedian=statsData.groupby(['Country Name','Indicator Name'])['value'].median().to_frame().reset_index() # meme chose avec median et ecart type
dfmedian=dfmedian.set_index(['Country Name','Indicator Name']).unstack()['value'].reset_index()

df=df.set_index(['Country Name','Indicator Name']).unstack()['value'].reset_index()
df.head(20)
```

```python
listIndicatorChosen
```

  ### <a name="C4">Indicateurs de croissance démographique</a>
 
 #### 1 croissance de la population

```python
# Fonction qui permet de creer un dataframe pour un indicateur donnée incluant la moyenne, l'écart type et la mediane
def dfMoySTDMed(indicatorName) :
    dfPG=df[['Country Name',indicatorName]]

    dfPGSTD=dfSTD[['Country Name',indicatorName]]
    dfPGSTD=dfPGSTD.rename(columns={indicatorName:'STD'} )


    dfPGmedian=dfmedian[['Country Name',indicatorName]]
    dfPGmedian=dfPGmedian.rename(columns={indicatorName:'median'} )


#jointure
    dfPG =dfPG.merge(dfPGSTD,on='Country Name',how='inner').merge(dfPGmedian,on='Country Name',how='inner')

    return dfPG
# Fonction qui permet de creer des graphes en bar avec barre d'erreur (= STD) par region
def GraphRegion(dfPG,indicatorName,title):

    dfPGR=dfPG.loc[dfPG['Country Name'].isin(listArea),:]
    dfPGR=dfPGR.sort_values(by=[indicatorName], ascending=False)

    plt.bar(height=dfPGR[indicatorName], x=dfPGR['Country Name'], yerr=dfPGR['STD'] )
    plt.xlabel("Région")
    plt.ylabel(indicatorName)
    plt.xticks(rotation = 'vertical')
    plt.title(title)
    plt.show()

    # Fonction qui permet de creer des graphes en bar avec barre d'erreur (= STD) par pays
def GraphCountry(dfPG,indicatorName,title):

    dfPGC=dfPG.loc[dfPG['Country Name'].isin(listCountryONU),:]
    dfPGC=dfPGC.sort_values(by=[indicatorName], ascending=False)
    dfPGC=dfPGC.iloc[:10]

    plt.bar(x=dfPGC['Country Name'],height=dfPGC[indicatorName], yerr=dfPGC['STD'])
    plt.xlabel("Pays")
    plt.ylabel(indicatorName)
    plt.xticks(rotation = 'vertical')
    plt.title(title)
    plt.show()
```

```python
dfPG=dfMoySTDMed('Population growth (annual %)')
dfPG.head(10)
```

```python
GraphRegion(dfPG,'Population growth (annual %)','Croissance de la population annuelle par région')
```

```python
GraphCountry(dfPG,'Population growth (annual %)','Croissance de la population annuelle par pays')
```

Au niveau de régions mondiale, le continant africain affiche une croissance elevée, suivi de prés des pays d'arabes et de sud asiatique. Nottons qu'au niveau des pays , les pays arabes de golf se positionnent aux sommet au niveau de taux de croissance démographique.
